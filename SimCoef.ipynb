```python
D1 = "Shipment of gold damaged in a fire"
D2 = "Delivery of silver arrived in a silver truck"
D3 = "Shipment of gold arrived in a truck"
Q = "gold silver truck"
```

```python
def freq(str):
    str = str.split()
    str2 = {}
    for i in str:             
        if i not in str2.keys():
            str2[i] = 1
        else:
            str2[i] = str2[i]+1
    return str2
```

```python
freqD1 = freq(D1)
freqD2 = freq(D2)
freqD3 = freq(D3)
print(freqD1)
print(freqD2)
print(freqD3)
```

```python
Total = D1+" "+D2+" "+D3
uniqueWords = []
Total = Total.split()
for i in Total:
    if i not in uniqueWords:
        uniqueWords.append(i)
print(uniqueWords)
```

```python
import math
def idfs(word):
    if(word in D1 and word in D2 and word in D3):
        return math.log(1,10)
    elif((word in D1 and word in D2) or (word in D2 and word in D3) or (word in D1 and word in D3)):
        return math.log(1.5,10)
    else:
        return math.log(3,10)
```

```python
idfList = {}
for i in uniqueWords:
    idfList[i] = float("{:.3f}".format(idfs(i)))
print(idfList)
```

```python
QList = Q.split()
QList
```

```python
freqQ = freq(Q)
freqQ
```

```python
simCoefD1 = 0
for word in freqQ.keys():
    if(word not in freqD1.keys()):
        continue
    else:
        simCoefD1 = simCoefD1 + freqD1[word]*idfList[word]*freqQ[word]*idfList[word]
print("Similarity coefficient of given query with D1 is ",simCoefD1)
```

```python
simCoefD2 = 0
for word in freqQ.keys():
    if(word not in freqD2.keys()):
        continue
    else:
        simCoefD2 = simCoefD2 + freqD2[word]*idfList[word]*freqQ[word]*idfList[word]
print("Similarity coefficient of given query with D2 is ",simCoefD2)
```

```python
simCoefD3 = 0
for word in freqQ.keys():
    if(word not in freqD3.keys()):
        continue
    else:
        simCoefD3 = simCoefD3 + freqD3[word]*idfList[word]*freqQ[word]*idfList[word]
print("Similarity coefficient of given query with D3 is ",simCoefD3)
```

```python
if(simCoefD1>simCoefD2):
    if(simCoefD1>simCoefD3):
        print("Q is much similar with D1 document")
    else:
        print("Q is much similar with D3 document")
else:
    if(simCoefD2>simCoefD3):
        print("Q is much similar with D2 document")
    else:
        print("Q is much similar with D3 document")
```
