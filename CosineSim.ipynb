```python
x = [3,2,0,5]
y = [1,0,0,0]
```

```python
inter = set(x).intersection(set(y))
union = set(x).union(set(y))
print(inter)
print(union)
```

```python
jacSim = len(inter)/len(union)
print("Jaccard Similarity is",jacSim)
```

Cosine Sim

```python
import math
up = 0
first = 0
second = 0
for i in range(len(x)):
    up = up + (x[i]*y[i])
    first = first+(x[i]*x[i])
    second = second+(y[i]*y[i])
first = math.sqrt(first)
second = math.sqrt(second)
cosSim = up/(first*second)
print("Cosine Similarity is",cosSim)
```
